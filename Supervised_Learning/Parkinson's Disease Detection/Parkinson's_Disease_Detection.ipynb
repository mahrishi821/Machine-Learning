Data Pre-processing

```python
#separting the features & Target
X=parkinsons_data.drop(columns=['name','status'],axis=1)
Y=parkinsons_data['status']
```

```python
print(X)
```

```python
print(Y)
```

```python
#Splitting the data in train and test data
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,random_state=3)
```

```python
print(X.shape,X_train.shape,X_test.shape)
```

```python
#Data Standarization
scaler=StandardScaler()
```

```python
scaler.fit(X_train)
```

```python
X_train=scaler.transform(X_train)
X_test=scaler.transform(X_test)
```

```python
print(X_train)
```

Model Training

```python
 #Support Vector Machin Model
 model=svm.SVC(kernel='linear')
```

```python
#training the svm model with training data
model.fit(X_train,Y_train)
```

Model Evaluation

```python
#accuracy score on traing data
X_train_prediction=model.predict(X_train)
training_data_accuracy=accuracy_score(Y_train,X_train_prediction)
print('Accuracy on the training Data is :: ',training_data_accuracy)
```

```python
 #accuracy score on testing data
X_test_prediction=model.predict(X_test)
testing_data_accuracy=accuracy_score(Y_test,X_test_prediction)
print('Accuracy on the testing Data is :: ',testing_data_accuracy)
```

Building a Predictive System
