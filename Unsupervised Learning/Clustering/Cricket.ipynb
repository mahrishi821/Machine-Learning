```python
import numpy as np
import pandas as pd
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
```

```python
df=pd.read_csv("/content/Cricket.csv",sep=",",encoding="ISO-8859-1",header=0)
df.head(100)
```

```python
features=['SR','Ave']
X=df[features]
```

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
kmeans=KMeans(n_clusters=4,random_state=100)
```

```python
kmeans.fit(X_scaled)
```

```python
df['cluster']=kmeans.labels_
```

```python
print("Cluster Centers:")
print(scaler.inverse_transform(kmeans.cluster_centers_))
```

```python
df['Player']=df['Player'].apply(lambda x:x.split('(')[0].strip())
```

```python
print(df['Player'].unique())
```

```python

virat_cluster=df.loc[df['Player']=='V Kohli','cluster'].values[0]
same_cluster_players=df[df['cluster']==virat_cluster]['Player'].tolist()
same_cluster_players.remove('V Kohli')

print("Players in the same cluster as Virat Kohli:")
print(same_cluster_players)

```
